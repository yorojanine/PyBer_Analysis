**5.1.3 Create Line Charts**

```python
%matplotlib inline
```

```python
#import dependencies
import matplotlib.pyplot as plt
```

```python
#set the x-axis to a list of strings for each month
x_axis = ["Jan", "Feb", "Mar", "April", "May", "June",
          "July", "Aug", "Sept", "Oct", "Nov", "Dec"]

#set the y-axis to a list of floats as the total fare 
#in US Dollars accumulated each month

y_axis = [
    10.02, 23.24, 39.20, 35.42, 32.34, 27.04,
    43.82, 10.56, 11.85, 27.90, 20.71, 20.09]
```

```python
#create the plot
plt.plot(x_axis, y_axis)
```

*Create a Line Chart Using the Object-Oriented Interface*

```python
#create the plot with ax.plt()
fig, ax = plt.subplots()
ax.plot(x_axis, y_axis)
```

```python
#second version of using ax.plt
fig = plt.figure()
ax = fig.add_subplot()
ax.plot(x_axis,y_axis)
```

```python
#quick graph using the object oriented interface method
ax = plt.axes()
ax.plot(x_axis,y_axis)
```

```python
#To change figure-level attributes (such as axis labels, a title, or a legend) 
#or save the figure as an image, use 
fig = plt.figure()
```

```python
#create the plot
plt.plot(x_axis, y_axis)
plt.show()
```

**5.1.4 Annotate Charts**

```python
# Create the plot and add a label for the legend.
plt.plot(x_axis, y_axis, 
        label='Boston',
        marker="*",
        color="blue",
        linewidth=2
        )
# Create labels for the x and y axes.
plt.xlabel("Date")
plt.ylabel("Fare($)")
# Set the y limit between 0 and 45.
plt.ylim(0,45)
# create a title
plt.title("PyBer Fare by Month")
#add a grid
plt.grid()
#add the lengend
plt.legend()
```

**5.1.7 Create Scatter Plots and Bubble Charts**

```python
#Create a Scatter Plot Using the MATLAB Method
#use the same data for x & y
# Set the x-axis to a list of strings for each month.
x_axis = ["Jan", "Feb", "Mar", "April", "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"]

# Set the y-axis to a list of floats as the total fare in US dollars accumulated for each month.
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]
```

```python
plt.plot(x_axis,y_axis,'o')
```

```python
plt.scatter(y_axis,x_axis, c = 'r', label = "Chicago")
plt.xlabel("Fare($)")
plt.xlim(0,50) #(min,max)
plt.ylabel("Date")
plt.gca().invert_yaxis()
plt.title("PyBer Fare by Month")
plt.legend()
```

```python
#Create a Bubble Chart Using the MATLAB Method
#Bubble charts are useful when presenting financial, population,
#and weather data because you can add a third and fourth factor to convey more information.

plt.scatter(x_axis,y_axis, s=y_axis)
```

```python
y_axis_larger = []
for data in y_axis:
    y_axis_larger.append(data*3)
```

```python
plt.scatter(x_axis,y_axis, s=y_axis_larger)
```

```python
plt.scatter(x_axis,y_axis, s = [i*3 for i in y_axis])
```

```python
fig, ax = plt.subplots()
ax.scatter(x_axis,y_axis)
```

```python
#Create a Bubble Chart Using the Object-Oriented Interface
fig, ax = plt.subplots()
ax.scatter(y_axis,x_axis, c="skyblue",
           s=[i*5 for i in y_axis],label="Boston"
          #,alpha = 0.2 #looks too light
           ,alpha = 0.8 #is it the opposite?
           ,edgecolors = 'black'
          ,linewidths = 2)
ax.set_xlim(0,50)
ax.invert_yaxis()
ax.set_xlabel("Fare($)")
ax.set_ylabel("Date")
ax.set_title("Pyber Fare by Month")
plt.legend()
```

**5.1.8 Create Pie Charts**

```python
#Create a Pie Chart Using the MATLAB Method
plt.pie(y_axis, labels= x_axis)
plt.show()
```

```python
plt.subplots(figsize=(8,8))
explode_values = (0,0,0,0,0,0,0.2,0,0,0,0,0)
plt.pie(y_axis,explode=explode_values,labels=x_axis, autopct='%1.1f%%')
```

```python
# Assign 12 colors, one for each month.
colors = ["slateblue", "magenta",
    "lightblue", "green",
    "yellowgreen", "greenyellow",
    "yellow", "orange",
    "gold", "indianred",
    "tomato", "mistyrose"]
explode_values = (0,0,0,0,0,0,0.2,0,0,0,0,0)
plt.subplots(figsize=(8,8))
plt.pie(y_axis,
       explode= explode_values,
       colors= colors,
       labels= x_axis,
    autopct='%1.1f%%')

plt.show()
```

**5.1.9 Chart Extras**

```python
# %matplotlib inline - already added
# import matplotlib.pyplot as plt - already added
import statistics
```

```python
# Set the x-axis to a list of strings for each month.
x_axis = ["Jan", "Feb", "Mar", "April", "May", "June", 
          "July", "Aug", "Sept", "Oct", "Nov", "Dec"]

# Set the y-axis to a list of floats as the total fare in 
#US dollars accumulated for each month.
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 
          43.82, 10.56, 11.85, 27.90, 20.71, 20.09]
```

```python
# Get the standard deviation of the values in the y-axis.
stdev = statistics.stdev(y_axis)
stdev
```

```python
plt.errorbar(x_axis,y_axis,yerr=stdev,
             capsize = 3)
```

```python
fig, ax = plt.subplots()
ax.errorbar(x_axis,y_axis, yerr=stdev, capsize=3)
plt.show()
```

```python
#change the major ticks
import numpy as np
plt.barh(x_axis,y_axis)
plt.xticks(np.arange(0,51,step=5.0))
plt.gca().invert_yaxis()
```

```python
fig, ax = plt.subplots()
ax.barh(x_axis,y_axis)
#ax.set_xticks(np.arange(0,51,step=5.0))
ax.xaxis.set_ticks(np.arange(0,51,step=5.0))
plt.show()
```

```python
#MINOR TICKS
from matplotlib.ticker import MultipleLocator
#increase the size of the plot figure
fig, ax = plt.subplots(figsize=(8,8))
ax.barh(x_axis,y_axis)
ax.set_xticks(np.arange(0,51, step =5.0))

#create minor ticks at an increment of 1
ax.xaxis.set_minor_locator(MultipleLocator(1))
plt.show()
```
